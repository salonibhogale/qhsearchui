```python
import pandas as pd
import os
import ndjson
```

```python
tar_folder = os.path.join('data')
tar_files = [i for i in os.listdir(tar_folder) if(i.endswith("ndjson"))]
```

```python
for tar_file in tar_files:
    with open(os.path.join(tar_folder, tar_file)) as f:
        data = ndjson.load(f)
        for row in data:
            if 'Q_Link' in row:
                row['Q_Link'] = row['Q_Link'].replace('http://164.100.47.194/Loksabha/', 'https://loksabha.nic.in/')


    with open(os.path.join(tar_folder, tar_file), 'w') as f:
        writer = ndjson.writer(f, ensure_ascii=False)

        for row in data:
            writer.writerow(row)
```
